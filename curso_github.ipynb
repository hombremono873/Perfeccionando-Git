# Ruta compacta para especializarme en GitHub (en orden)

## 1. Fundamentos sólidos de Git
- working directory, staging area, repository
- commit, HEAD, historial
- git add / commit / status / log / diff / restore

## 2. Trabajo con repositorios remotos
- git clone
- git remote / origin / upstream
- git fetch vs git pull
- git push
- resolución básica de conflictos

## 3. Flujo con ramas (branching profesional)
- main y develop
- ramas feature/*
- git switch / branch
- git merge
- rebase (nivel intermedio)
- eliminación de ramas

## 4. Pull Requests (flujo real de trabajo)
- crear PR desde una rama
- revisar cambios
- comentar código
- aprobar y hacer merge
- cerrar issues desde el PR

## 5. Issues para seguimiento
- crear tareas y bugs como issues
- etiquetas (labels)
- asignaciones
- relacionar commits con issues
- cerrar issues con mensajes de commit

## 6. GitHub Projects (tableros de trabajo)
- columnas: backlog / todo / doing / done
- vincular issues al tablero
- seguimiento del avance del proyecto

## 7. Buenas prácticas de repositorio
- README profesional
- estructura clara de carpetas
- commits pequeños y con mensajes claros
- uso correcto de .gitignore
- versionado por hitos (v0.1, v0.2, v1.0, etc.)

## 8. Flujo personal como si fuera un equipo
- trabajar siempre desde ramas
- nunca desarrollar directamente en main
- todo cambio entra por Pull Request
- cada funcionalidad = una rama

## 9. Gestión de entregables y versiones
- tags y releases
- changelog básico
- control de versiones de entregas académicas o técnicas

## 10. Documentación del proyecto
- descripción del problema
- cómo instalar
- cómo ejecutar
- dependencias
- estructura del proyecto

## 11. Automatización básica (GitHub Actions – opcional avanzado)
- ejecutar tests
- validar que el proyecto corre
- chequeos automáticos en Pull Requests

## 12. Portafolio profesional en GitHub
- repositorios completos (no scripts sueltos)
- historial limpio de commits
- uso visible de ramas, PR e issues
- proyectos con contexto real

## 13. Perfil GitHub
- definir pipelines
- automatizar pruebas
- ejecutar validaciones
- hacer despliegues
